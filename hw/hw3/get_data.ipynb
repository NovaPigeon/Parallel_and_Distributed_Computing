```python
import re
import pandas as pd
mpi_data=[]
seq_data=[]
with open('./mpi.log') as mpi_res:
    for line in mpi_res:
        numbers = re.findall(r'\d+\.*\d*', line)
        mpi_data.append(numbers)
with open('./seq.log') as seq_res:
    for line in seq_res:
        numbers = re.findall(r'\d+\.*\d*', line)
        seq_data.append(numbers)
```

```python
# 将数据转换为 DataFrame
df_mpi = pd.DataFrame(mpi_data, columns=['Matrix N', 'MPI Execution time', 'Worker num'])
df_seq= pd.DataFrame(seq_data, columns=['Matrix N', 'Sequence Execution time'])

# 将字符串列转换为数值型
df_mpi['Matrix N'] = pd.to_numeric(df_mpi['Matrix N'])
df_mpi['MPI Execution time'] = pd.to_numeric(df_mpi['MPI Execution time'])
df_mpi['Worker num'] = pd.to_numeric(df_mpi['Worker num'])

df_seq['Matrix N'] = pd.to_numeric(df_seq['Matrix N'])
df_seq['Sequence Execution time'] = pd.to_numeric(df_seq['Sequence Execution time'])

# 使用 pivot 方法生成二维表格
table_mpi = df_mpi.pivot(index='Matrix N', columns='Worker num', values='MPI Execution time')
df_seq.set_index('Matrix N',inplace=True)
table_seq=df_seq

table_ratio=pd.DataFrame(table_seq.min(axis=1)/table_mpi.min(axis=1),columns=['Ratio of Seq/MPI'])
```
